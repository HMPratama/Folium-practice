```python
# Peta Melbourne
melb_map = folium.Map(location = [-37.93168, 145.16126],
                     zoom_start=10)
# buat Marker Liberty Statue
mansion = folium.Marker(location = [-37.93168, 145.16126],
                        # popup => Ketika marker di klik maka akan muncul kalimat sesuai dengan di popul
                        popup = folium.Popup('35 Bevis St, Mulgrave VIC 3170, Australia', max_width=500 ),
                        # tooltip => ketika cursor di hover ke marker
                        tooltip ="Mulgrave VIC 3170, Australia",
                        icon = folium.Icon('purple',icon="home",prefix="glyphicon",))
mansion.add_to(melb_map)
melb_map
```

> ### **Insight:**

### **Nomor 2**

- Ceritanya kita sedang mencari apartment (type='u') dengan 2 kamar tidur (Bedroom2) dengan harga yg murah
- Coba tampilkan peta kota Melbourne
- Berikan marker berupa pin pada 5 apartment termurah
- Berikan tooltip berupa harga rumah
- Berikan popup berupa Suburb dan Address

```python
df[['Bedroom2']]
```

```python
df_apt=df[(df['Type']=='u')&(df['Bedroom2']==2)].sort_values('Price')
df_apt[['Suburb','Address','Lattitude','Longtitude']]
df_apt.head()
```

```python
# Buat list
prc = list(df_apt.Price)
sub = list(df_apt.Suburb)
ade = list(df_apt.Address)
latid = list(df_apt.Lattitude)
lont = list(df_apt.Longtitude)
```

```python
melb_map = folium.Map(location = [-37.969643, 144.3937186],
                     zoom_start=8)
for i in range(6):
    mapmarker = folium.Marker(location=[latid[i],lont[i]],
                              popup = folium.Popup(f'{sub[i]},{ade[i]}',max_width=500),
                              tooltip =  f'{prc[i],sub[i]}',
                              icon = folium.Icon(color="red",icon="home", prefix='fa'))
    mapmarker.add_to(melb_map)
melb_map
```

>### **Insight:**

>> Dari data untuk apartement 2 bedroom dengan harga yang murah ada 5 apartement di daerah Melbourne.
